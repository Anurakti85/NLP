```
#Tokenizing the sentences with tensorflow 
```

```
from tensorflow.keras.preprocessing.text import Tokenizer
```

```
sentences = [
  'I love my country',
  'I like yellow flowers',
  'Did you lives in England?'
]

tokenizer = Tokenizer(num_words = 100, oov_token="<OOV>")
tokenizer.fit_on_texts(sentences)
indexed_words = tokenizer.word_index

print(indexed_words)
```

```
#Code for Sequencing
```

```
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```
sequences = tokenizer.texts_to_sequences(sentences)

padded = pad_sequences(sequences, maxlen=5)
print("\nThe Word Index = " , indexed_words)
print("\nThe Sequences = " , sequences)
print("\nThe Padded Sequences:")
print(padded)


# Test data with sentences we have never seen 
test_data = [
    'i am from England',
    'you love to travel around the whole world'
]

test_seq = tokenizer.texts_to_sequences(test_data)
print("\nTest Sequence = ", test_seq)

padded = pad_sequences(test_seq)
print("\nPadded Test Sequence: ")
print(padded)
```
