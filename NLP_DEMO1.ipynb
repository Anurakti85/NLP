```
#Term Frequency Example
#importing required libraries
import nltk
nltk.download(("stopwords"))
from nltk.probability import FreqDist
from nltk.corpus import stopwords

```

```
#Stopwords removal
stop_words = set(stopwords.words('english'))
```

```
#Reading the Corpus
nltk.download('gutenberg')
words=nltk.Text(nltk.corpus.gutenberg.words('bryant-stories.txt'))
```

```
#printing unique words
print(len(set(words)))
```

```
print(words)
```

```
#changing all corpus to small letters
words=[word.lower() for word in words if word.isalpha()]
words=[word.lower() for word in words if word not in stop_words]
```

```
words
```

```
#creating an object of FreqDist class and passing the  cleaned corpus into it.
fdist=FreqDist(words)
fdist
```

```
print(len(words))
```

```
#Raw count of words
for x,v in fdist.most_common(10): 
     print(x,v)
```

```
#Term frequency adjusted to document length
for x,v in fdist.most_common(10): 
     print(x,v/len(fdist))
```
